```python
from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup
class WebCrawler:
    
    def __init__(self):
        self.options = webdriver.ChromeOptions()
        self.options.add_argument("headless")
        self.options.add_experimental_option('excludeSwitches', ['enable-logging'])
        self.driver = webdriver.Chrome(options=self.options)
    

    def fetch_page(self,url):
        self.driver.get(url)
        soup = BeautifulSoup(self.driver.page_source, 'html.parser')
        return soup
            
```

```python
url_page = 'https://www.sreality.cz/en/search/for-sale/apartments'
#https://www.sreality.cz/en/search/for-sale/apartments?page=2
web_crawler = WebCrawler()
soup = web_crawler.fetch_page(url_page)
soup
```

```python
#first page is fetched without ?page={number_of_page} every other page is fetched in that way
all_apartments = soup.find_all(class_ = 'property ng-scope')
pages_to_fetch = 25
#pages_to_fetch + 1 because range is exclusive
for page_number in range(2, 500):
    query_param = '?page=' + str(page_number)
    soup = web_crawler.fetch_page(url_page + query_param)
    apartments = soup.find_all(class_ = 'property ng-scope')
    all_apartments.extend(apartments)
    print(len(all_apartments))
    if len(all_apartments) == 500:
        break

len(all_apartments)
```
